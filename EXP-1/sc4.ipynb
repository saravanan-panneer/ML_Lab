```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Saravanan 24BAD105")
#1. Import Dataset

df = pd.read_csv("marketing_campaign.csv", sep="\t")
print("Dataset successfully loaded!\n")

# 2. Examine Structure
print("First 5 rows of the dataset:")
print(df.head(), "\n")

print("Dataset Information:")
print(df.info(), "\n")

print("Null Values in Each Column:")
print(df.isnull().sum(), "\n")

# 3. Feature Engineering
# Create Age column
df["Age"] = 2025 - df["Year_Birth"]

# 4. Visualizations
sns.set(style="whitegrid")

# Age Distribution (Bar Plot)
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

# Income Distribution (Box Plot)
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Income"])
plt.title("Customer Income Distribution")
plt.xlabel("Income")
plt.show()

# Spending Patterns (Box Plot)
spending_columns = [
    "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"
]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[spending_columns])
plt.title("Customer Spending Patterns")
plt.ylabel("Amount Spent")
plt.xticks(rotation=45)
plt.show()


# 5. Summary Insights
print("INSIGHTS FROM VISUALIZATIONS:")
print("1. Most customers belong to the middle-age group.")
print("2. Income data contains missing values and several outliers.")
print("3. Higher spending is observed on wine and meat products.")
print("4. Some spending categories show low median values, indicating selective purchases.")
```
